```
import base64
import io
import re
import timeit
import uuid
import ast

import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display

from google.cloud import aiplatform
from google.cloud import storage
from google.cloud.aiplatform.gapic import PredictResponse


# Parses the generated image bytes from the response and converts it
# to PIL Image object.
def prediction_to_pil_image(prediction: PredictResponse, size=(640,640)) -> Image.Image:
    encoded_bytes_string = prediction["bytesBase64Encoded"]
    decoded_image_bytes = base64.b64decode(encoded_bytes_string)
    image_pil = Image.open(io.BytesIO(decoded_image_bytes))
    image_pil.thumbnail(size)
    return image_pil


# Displays PIL images in a row horizontally.
def display_row(input_images: list, figsize: tuple[int, int] = (15, 15)):
    count = len(input_images)
    fig, ax = plt.subplots(1, count, figsize=figsize)

    if count == 1:
      ax.imshow(input_images[0], None)
      ax.axis("off")
    else:
      for i in range(count):
        ax[i].imshow(input_images[i], None)
        ax[i].axis("off")

    plt.show()

# Download image bytes from a GCS URI.
def download_gcs_image_bytes(uri:str) -> bytes:
  match = re.match(r"gs://(.*?)/(.*)", uri)
  if not match:
      print(f"Invalid GCS URI format: {uri}")

  bucket_name = match.group(1)
  object_name = match.group(2)

  storage_client = storage.Client()
  bucket = storage_client.bucket(bucket_name)
  blob = bucket.blob(object_name)
  return blob.download_as_bytes()


# Constructs a Vertex AI PredictRequest and uses it to call Imagen Product Recontext
def call_product_recontext(
    image_bytes_list: list = [],
    image_uris_list: list = [],
    prompt: str = None,
    product_description: str = None,
    disable_prompt_enhancement: bool = False,
    sample_count: int = 1,
    storage_uri: str = None, # Add the storage_uri argument
) -> PredictResponse:
    instances = []

    instance = {
        "productImages": []
    }

    for product_image_bytes in image_bytes_list:
        product_image = {
            "image": {
                "bytesBase64Encoded": product_image_bytes
            }
        }
        instance["productImages"].append(product_image)

    for product_image_uri in image_uris_list:
        product_image = {
            "image": {
                "gcsUri": product_image_uri
            }
        }
        instance["productImages"].append(product_image)

    if product_description:
        instance["productImages"][0]["productConfig"] = {
            "productDescription": product_description
        }

    if prompt:
        instance["prompt"] = prompt

    parameters = {}

    # Add sample_count to the parameters dictionary
    if sample_count > 0:
        parameters["sampleCount"] = sample_count

    # Add storageUri to the parameters dictionary if provided
    if storage_uri:
        parameters["storageUri"] = storage_uri

    if disable_prompt_enhancement:
        parameters["enhancePrompt"] = False

    instances.append(instance)

    start = timeit.default_timer()

    response = client.predict(
        endpoint=model_endpoint, instances=instances, parameters=parameters
    )
    end = timeit.default_timer()
    print(f"Product Recontextualization took {end - start:.2f}s.")

    return response
```

```
# @title Set product images
# @markdown Run this cell to enable and select the Choose files button.
# @markdown You can then select one to three images file from your local device to upload.
# @markdown Large images are resized to a maximum dimension of 1024 pixels for faster processing.

images = files.upload()
image_bytes_list = []
image_pils = []

num_images = min(3, len(images.values()))
for i in range(num_images):
  RAW_IMAGE_BYTES = list(images.values())[i]
  ENCODED_IMAGE_BYTES = base64.b64encode(RAW_IMAGE_BYTES).decode("utf-8")
  image_bytes_list.append(ENCODED_IMAGE_BYTES)

  PRODUCT_IMAGE_PIL = Image.open(io.BytesIO(RAW_IMAGE_BYTES)).convert("RGB")
  image_pils.append(PRODUCT_IMAGE_PIL)
  PRODUCT_IMAGE_PIL.thumbnail((1024, 1024))
  plt.axis("off")
  plt.imshow(PRODUCT_IMAGE_PIL)

  width, height = PRODUCT_IMAGE_PIL.size
  print(f"image size(width x height): {width} x {height}")
```

#### Send request with a prompt

Set a `prompt` in the request to describe how the generated image should look.

You can also set a `product_description` as a short description of the product itself.

```
# Define parameters

product_description = "banner" # @param {'type': 'string'}
prompt = "\"Happy birthday\" banner in a birthday party with Dimensions 7.10 x 5.90 x 0.40 Inches, high-quality product photography aesthetic, Commercial Lifestyle Color Grading" # @param {'type': 'string'}
disable_prompt_enhancement = False # @param {'type': 'boolean'}
NUMBER_OF_IMAGES = 4
OUTPUT_GCS_URI = "gs://sandbox-401718-visual-cat-enrichment/celebration-journey" # @param {'type': 'string'}


# Call the API

response = call_product_recontext(
    prompt=prompt,
    image_bytes_list=image_bytes_list,
    product_description=product_description,
    disable_prompt_enhancement=disable_prompt_enhancement,
    sample_count=NUMBER_OF_IMAGES,
    storage_uri=OUTPUT_GCS_URI # Pass the storage URI here
)


# Generate and display the images

generated_images = []
print("Processing predictions...")

for pred in response.predictions:
    if "gcsUri" in pred:
        output_uri = pred["gcsUri"]
        print(f"Image was saved to: {output_uri}")
        image_bytes = download_gcs_image_bytes(output_uri)
        image_pil = Image.open(io.BytesIO(image_bytes))
        generated_images.append(image_pil)
    elif "bytesBase64Encoded" in pred:
        print("Image was returned as base64 bytes.")
        generated_images.append(prediction_to_pil_image(pred))

# Display all the generated images
display_row(generated_images)
```

## Option B: Run Async Image Generation and iterate over rows with defined parameters

Paramaters include Product Title, Prompt, Product Path URIs, and	Output Bucket

```
# Define Function

import asyncio
import time

async def process_one_row_async(index, row, semaphore, number_of_images):
    """
    This is the async worker. It processes one row from the DataFrame.
    It waits for an open slot from the semaphore, then runs the API call in a
    separate thread to avoid blocking the main event loop.
    """
    async with semaphore:
        try:
            # Prepare data
            product_title = row['Product Title']
            prompt = row['Prompt']
            output_bucket_base = row['Output Bucket']
            uris_string = row['Product Path URIs']
            image_uris_list = ast.literal_eval(uris_string)

            safe_title = re.sub(r'[^a-zA-Z0-9_-]', '', product_title.replace(' ', '_'))
            unique_id = str(uuid.uuid4())[:8]
            output_storage_uri = f"{output_bucket_base.rstrip('/')}/{safe_title}_{unique_id}/"

            print(f"-> Starting job for Row {index}: '{product_title}'")

            # Call the synchronous function
            # call_product_recontext function without freezing the whole program
            response = await asyncio.to_thread(
                call_product_recontext,
                prompt=prompt,
                image_uris_list=image_uris_list,
                product_description=product_title,
                disable_prompt_enhancement=False,
                sample_count=number_of_images,
                storage_uri=output_storage_uri
            )

            # Process the result
            generated_uris = [pred["gcsUri"] for pred in response.predictions if "gcsUri" in pred]
            print(f"Finished job for Row {index}: '{product_title}' ({len(generated_uris)} images)")

            return {
                "product_title": product_title,
                "status": "Success",
                "output_uris": generated_uris
            }

        except Exception as e:
            product_title = row.get('Product Title', f'Row {index}')
            print(f"[ERROR] for '{product_title}': {e}")
            return {
                "product_title": product_title,
                "status": "Failed",
                "error": str(e)
            }
```

```
# EXAMPLE INPUT DATA TABLE
csv_data = """Product Title,Prompt,Product Path URIs,Output Bucket,Metadata
Wine bottle,"Wine bottle titled slightly in a vineyard with bright lighting, high-quality product photography aesthetic, Commercial Lifestyle Color Grading","['gs://sandbox-401718-visual-cat-enrichment/celebration-journey/input/Woodbridge-Cabernet-Sauvignon-Red-Wine-1-5-L-Bottle-13-5-ABV_7615a357-13ff-4175-9bae-19f8ace98efe.32dcd04f82ecc0d0f01b867da4443dc1.jpg']",gs://sandbox-401718-visual-cat-enrichment/celebration-journey/output,"{journey: Celebration Journey}"
Flower Bouquet,"Bouquet of flowers displayed on a table of a fancy dinner party, high-quality product photography aesthetic, Commercial Lifestyle Color Grading","['gs://sandbox-401718-visual-cat-enrichment/celebration-journey/input/Fresh-Cut-Extra-Large-Mixed-Flower-Bouquet-Minimum-of-17-Stems-Colors-Vary_5158bbdc-5d43-4313-867f-4dd4f2168e1b.fe0a6fa9fd5adf881e4d87de042e15b9.jpg']",gs://sandbox-401718-visual-cat-enrichment/celebration-journey/output,"{journey: Celebration Journey}"
"""
df = pd.read_csv(io.StringIO(csv_data))
display(df)
```

```
NUMBER_OF_IMAGES_PER_ROW = 2  # @param
CONCURRENT_REQUESTS = 5       # @param

# Orchestration Logic
# A Semaphore limits how many tasks can run at the same time.
semaphore = asyncio.Semaphore(CONCURRENT_REQUESTS)
tasks = []

# list of tasks, one for each row in the DataFrame.
for index, row in df.iterrows():
    task = process_one_row_async(index, row, semaphore, NUMBER_OF_IMAGES_PER_ROW)
    tasks.append(task)

print(f"Starting batch processing for {len(tasks)} rows with concurrency={CONCURRENT_REQUESTS}...")
start_time = time.time()

# `asyncio.gather` runs all tasks concurrently and collects their results.
# The list of results will be in the same order as the DataFrame rows.
list_of_results = await asyncio.gather(*tasks)

end_time = time.time()
print(f"\n--- Batch processing complete ---")
print(f"Total time: {end_time - start_time:.2f} seconds")

# Convert the list of result dictionaries into a new DataFrame
results_df = pd.DataFrame(list_of_results)
df_final = pd.concat([df, results_df], axis=1)

# Remove duplicate column
if 'product_title' in df_final.columns:
    df_final = df_final.loc[:, ~df_final.columns.duplicated()]

print("\nBatch Job Complete. Results have been appended to the DataFrame:")
display(df_final)
```
